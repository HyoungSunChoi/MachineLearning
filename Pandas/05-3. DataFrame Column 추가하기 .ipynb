## 학습목표
 1. Dataframe boolean selection 이해하기

```python
import pandas as pd
```

```python
train_data=pd.read_csv('./train.csv')
```

```python
train_data.head()
```

#### **boolean selection으로 row 선택하기**
 - numpy에서와 동일한 방식으로 해당 조건에 맞는 row만 선택

#### 예시. ) 30대 이면서 1등석에 탄 사람 선택하기

```python
class_= train_data['Pclass']==1
age_= (train_data['Age']>=30) & (train_data['Age']<40)

train_data[class_ & age_]
```

---

### DataFrame 에 새 컬럼 추가 / 삭제하기

#### 새 column 추가하기
 - [] 사용하여 추가하기
 - insert 함수 사용하여 원하는 위치에 추가하기

```python
train_data['Age_double']=train_data['Age']*2
train_data.head()
```

```python
train_data['Age_tripple']=train_data['Age_double']+train_data['Age']
train_data.head()
```

```python
train_data.insert(3,'Fare10',train_data['Fare']/10)
train_data.head()
```

---

#### column 삭제하기
 - drop 함수 사용하여 삭제
   - 리스트를 사용하여 멀티플 삭제 가능 

```python
train_data.drop('Age_tripple',axis=1)
train_data.head()
# 원본의 복사본을 삭제하는 것임!!!
```

```python
train_data.drop(['Age_tripple','Age_double'],axis=1)
train_data.head()
```

```python
train_data.drop(['Age_double','Age_tripple'],axis=1,inplace=True)
train_data.head()
# 원본 삭제는 inplace =  True
```
