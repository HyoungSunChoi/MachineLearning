#### column 선택하기
  - 기본적으로 [ ]는 column을 추출 
  - 컬럼 인덱스일 경우 인덱스의 리스트 사용 가능
    - 리스트를 전달할 경우 결과는 Dataframe
    - 하나의 컬럼명을 전달할 경우 결과는 Series

```python
import pandas as pd
import numpy as np
```

```python
train_data=pd.read_csv('./train.csv')
train_data.head()
```

#### 특정 컬럼만 가져올 때는 [ ]

```python
train_data['Pclass']
```

```python
train_data[['Survived','Pclass']]
```

---

#### dataframe slicing
  - dataframe의 경우 기본적으로 [] 연산자가 **column 선택**에 사용
  - 하지만, slicing은 row 레벨로 지원

```python
train_data[:3]
```

#### row 선택하기
  - Seires의 경우 []로 row 선택이 가능하나, **DataFrame의 경우는 기본적으로 column을 선택하도록 설계**
  - **.loc, .iloc함수**로 row 선택 가능
    - loc - 인덱스 자체를 사용
    - iloc - 0 based index로 사용
    - 이 두 함수는 ,를 사용하여 column 선택도 가능

```python
train_data.info()
```

```python
train_data.index=np.arange(100,991)
```

```python
train_data.tail()
```

하나만 선택하면 Series 형태로 나온다

```python
train_data.loc[986]
```

여러개는  list 형태로 받기

```python
train_data.loc[[100,110,660]]
```

iloc 0 based index

```python
train_data.iloc[0]
```

#### row column 동시에 선택하기
- loc, iloc 속성 이용시 콤마를 이용하여 둘다 사용!!!

```python
train_data.loc[[986,100,110,990],['Survived','Pclass','Name','Sex','Age']]
```

```python
train_data.iloc[[101,100,200,120],[1,4,5]]
# iloc 컬럼 가져올때 사용
```
