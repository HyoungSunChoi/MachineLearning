#### import
- import를 사용하여 해당 모듈 전체를 import

```python
import math
math.pi
```

```python
math.cos(100)
```

#### from import
- 해당 모듈에서 특정한 타입만 import 

```python
from math import pi
```

```python
pi
```

```python
from math import cos
```

```python
cos(100)
```

#### * 임포트
- 해당 모듈 내 모든 것은 import
- 일반적으로 사용이 권장되지 않음
    - 이름이 중복이 있을 수 있기 때문 (오류 발생)

```python
from math import *
```

```python
sin(180)
```

#### as
    - alias (별명) 지정가능


```python
import math as m
m.exp(3)
```
