# Get , Post

- Get 방식: 데이터를 URL 에 포함하여 전달 ( 리소스 요청에 사용됨 )

- Post 방식 : 데이터를 Form data 에 포함하여 전달 ( 로그인에 사용 )

### requests 모듈 사용하여 http request/response 확인하기

#### requests 모듈
1. http request/response를 위한 모듈
2. HTTP method를 메소드 명으로 사용하여 request 요청 예) get, post

```python
import requests
url = 'https://news.v.daum.net/v/20190728165812603'
```

#### get 요청하기
1. http get 요청하기
2. query parameter 이용하여 데이터 전달하기

```python
url = 'https://news.v.daum.net/v/20190728165812603'
resp=requests.get(url)
resp # 200 이면 정상적으로 크롤링한것!
resp.text # 사이트 정보 불러오기
```

#### post 요청하기
1. http post 요청하기
2. post data 이용하여 데이터 전달하기

```python
url='https://www.kangcom.com/member/member_check.asp'
headers={
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.141 Safari/537.36'
}
data = {
    'id':'macmath22',
    'pwd':'Test1357!'
}
resp=requests.post(url, data=data)
resp.text
```

#### HTTP header 데이터 이용하기
1. header 데이터 구성하기
2. header 데이터 전달하기

#### HTTP response 처리하기
1. response 객체의 이해
2. status_code 확인하기
3. text 속성 확인하기

```python
url='https://news.v.daum.net'
resp=requests.get(url)
resp

if resp.status_code==200:
    print(resp.headers)
else:
    print('error')
```
