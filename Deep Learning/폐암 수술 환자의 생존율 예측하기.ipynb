```python
from keras.models import Sequential
from keras.layers import Dense

import numpy
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

seed=0
numpy.random.seed(seed)
tf.set_random_seed(seed)

Data_set=numpy.loadtxt("./dataset/ThoraricSurgery.csv", delimiter=",")

# 환자의 기록과 수술 결과 X와 Y 로 저장 
X=Data_set[:, 0:17] # 0 ~ 16번째 열
Y=Data_set[:,17]

# 딥러닝 구조 결정 ( 모델 설정하고 실행하는 부분)
model=Sequential()
model.add(Dense(30, input_dim=17, activation='relu'))
    # 입력층과 은닉층
    # 데이터에서 17개의 값을 받아 은닉층의 30개 노드로 보낸다.
model.add(Dense(1, activation='sigmoid'))
    # 출력층
    # 출력값을 1개로 보내야하기 때문에 노드는 1개이다. 

# 딥러닝 실행
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
    # 어떤 오차함수를 사용할지 정해야 한다.
    # 여기서는 평균 제곱 오차 함수를 사용함 (mean_squared_error)
    # 최적화를 위해 adam 사용
    # metrics 함수를 사용
        # 정확도를 측정하기 위해 사용되는 테스트 샘플을 학습 과정에서 제외시켜 
        # 과적합 문제를 방지
        
# 모델 실행하기
model.fit(X, Y, epochs=30, batch_size=10)
# 총 470명의 환자에게서 17개의 정보를 정리한 것
# 학습 프로세스가 모든 샘플에 대해 한 번 실행되는 것을 1 epoch
    # epoch = 30 : 각 샘플을 처음부터 끝까지 30번 재사용될 때 까지 반복
# batch_size 는 샘플을 몇개씩 처리할지 정하는 것
    # 너무 클 경우 학습 속도가 느리다.
    # 너무 작으면 각 실행 값의 편차가 생겨,, 전체 결과값이 불안정해진다
    # TIP : 컴퓨터 메모리가 감당 가능한 값을 찾아서 설정할 것

# 결과 출력
print('\n Accuracy: %.4f'%(model.evaluate(X,Y)[1]))
```
