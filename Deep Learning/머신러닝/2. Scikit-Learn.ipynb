## Scikit-Learn
- Scikit-Learn 으로부터 적절한 추정기 클래스를 임포트해서 모델의 클래스를 선택
- 이 클래스를 원하는 값으로 인스턴스화 해서 모델의 초모수를 선택
- 데이터를 앞에서 논의한 내용에 따라 특징 배열과 대상 벡터로 배치
- 모델 인스턴스의 fit() 메서드를 호출해 모델을 데이터에 적합
- 모델을 새 데이터에 적용
    - 지도 학습은 predict()
    - 비지도 학습은 transform() 이나 predict() 메서드 활용

```python
import matplotlib.pyplot as plt
import numpy as np
rng = np.random.RandomState(42)
x= 10 * rng.rand(50)
y = 2*x-1 + rng.randn(50)
plt.scatter(x,y)
```

```python
# 1. 모델 클래스를 선택하기
from sklearn.linear_model import LinearRegression

# 2. 모델 초모수 선택하기
model = LinearRegression(fit_intercept=True)
model

# 3. 데이터를 특징 행렬과 대상 벡터로 배치한다.
X = x[:,np.newaxis]
X.shape

# 4. 모델을 데이터에 적합하기
model.fit(X,y)
# 모델의 기울기
model.coef_
# 모델의 절편
model.intercept_

# 5. 알려지지않은 데이터에 대한 레이블을 예측하기
xfit = np.linspace(-1,11)
Xfit = xfit[:, np.newaxis]
yfit = model.predict(Xfit)
plt.scatter(x,y)
plt.plot(xfit,yfit)
```
