### 퍼셉트론과 뉴런의 비교



y = ax + b
    - a: 기울기
    - b: y 절편
y = wx + b
    - w: 가중치
    - b: 바이어스
        * 가중합 : 가중치(w) * 값(x) 들의 합
        * 가중합의 결과를 1,0 을 판단하는 함수 -> 활성화 함수(시그모이드)

#### XOR 문제 해결하기

```python
import numpy as np
w11 = np.array([-2,-2])
w12 = np.array([2,2])
w2 = np.array([1,1])
b1=3
b2=-1
b3=-1

# 퍼셉트론
def MLP(x, w, b):
    y=np.sum(w*x)+b
    if y<=0:
        return 0
    else:
        return 1
    
# NAND 게이트
def NAND(x1,x2):
    return MLP(np.array([x1,x2]), w11, b1)

# OR 게이트
def OR(x1, x2):
    return MLP(np.array([x1,x2]), w12, b2)

# AND 게이트
def AND(x1, x2):
    return MLP(np.array([x1,x2]), w2, b3)

# XOR 게이트
def XOR(x1, x2):
    return AND(NAND(x1,x2), OR(x1,x2))

if __name__=="__main__":
    for x in [(0,0),(1,0),(0,1),(1,1)]:
        y=XOR(x[0],x[1])
        print('입력 값 :'+str(x) + '출력 값 : '+str(y))
```
