## 학습목표
 1. matplotlib 모듈 이용하여 그래프 표현하기

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
# jupyter 이용시 꼭 추가해주기
```

## 그래프 데이터 생성

```python
x=np.linspace(0,10,11)
y=x**2 + x + 2 + np.random.randn(11)
```

```python
print(x)
print(y)
```

### 그래프 출력하기
 - plot함수 (선 그래프), scatter(점 그래프), hist(히스토그램) 등 사용
   - 함수의 parameter 혹은 plt의 다른 함수로 그래프 형태 및 설정을 변경 가능
   - 기본적으로, x, y에 해당하는 값이 필요

```python
plt.plot(x,y)
```

```python
plt.scatter(x,y)
```

### 그래프에 주석 추가 

* **x, y 축 및 타이틀**

```python
plt.xlabel('X value')
plt.ylabel('Y value')
plt.title('X-Y rel')
plt.plot(x,y)
```

* **grid 추가** (선추가)

```python
plt.xlabel('X value')
plt.ylabel('Y value')
plt.title('X-Y rel')
plt.grid(True)
plt.plot(x,y)
```

* **x, y축 범위 지정**

```python
plt.xlabel('X value')
plt.ylabel('Y value')
plt.title('X-Y rel')
plt.grid(True)

plt.xlim(0,5)
plt.ylim(0,100)

plt.plot(x,y)
```

### plot 함수 parameters
 - 그래프의 형태에 대한 제어 가능
 - [plot함수 도큐먼트](https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot)
 - 원하는 스타일은 위에서 찾아보쟝

* 그래프의 색상 변경

```python
plt.plot(x,y,'#aa64ff')
# 색깔의 줄임말을 입력
# r : red , b : blue , y : yellow , k : black
# RGB 색상 사용 가능 : #ff00ff
```

* 그래프 선스타일 변경

```python
plt.plot(x,y,'-.')
```

```python
plt.plot(x,y,'g^')
```

```python
plt.plot(x,y,'r^')
```

```python
plt.plot(x,y,'m:')
```

* 그래프 두께 변경
 - linewidth 명시

```python
plt.plot(x,y,'m:',linewidth=9)
```

* keyword parameter 이용하여 모든 속성 설정
 - https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot
 - color
 - linestyle
 - marker
 - markerfacecolor
 - markersize 등등

```python
plt.plot(x,y,color='black',
         linestyle='--',
         marker='^',
         markerfacecolor='blue',
         markersize=20)
```

### subplot으로 여러 그래프 출력하기
 - subplot함수로 구획을 구별하여 각각의 subplot에 그래프 출력
     - 여러 가지 속성을 표현할 때 사용한다

```python
# 구역 설정하기  총 2*2 개수 중 ? 번째 그래프
plt.subplot(2,2,1)
plt.plot(x,y,'r')

plt.subplot(2,2,2)
plt.plot(x,y,'g')

plt.subplot(2,2,3)
plt.plot(y,x,'k')

plt.subplot(2,2,4)
plt.plot(x,np.exp(x),'b')
```

### hist함수 사용 (막대그램)
 - histogram 생성
 - bins로 historgram bar 개수 설정

```python
data=np.random.randint(1,100,size=200)
```

```python
# bins 는 빈도수 alpha 는 투명도
plt.hist(data,bins=20, alpha=0.3)
plt.xlabel('값')
plt.ylabel('개수')
plt.grid(True)
```
